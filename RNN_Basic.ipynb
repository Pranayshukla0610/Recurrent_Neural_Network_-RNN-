```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5],
    [4,5,6],
    [5,6,7],
    [6,7,8]
])

y = np.array([
    4,
    5,
    6,
    7,
    8,
    9
])
```

```python
print(X)
print(y)
```

```python
X = X.reshape((6,3,1))
print(X.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(
    SimpleRNN(
        units=10,
        input_shape=(3,1)
    )
)

model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss = 'mse'
)
```

```python
history = model.fit(
    X,
    y,
    epochs=500,
    verbose=1
)
```

```python
test = np.array([[7,8,9]])

test = test.reshape((1,3,1))

prediction = model.predict(test)

print(prediction)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.title("Training Loss")

plt.show()
```
